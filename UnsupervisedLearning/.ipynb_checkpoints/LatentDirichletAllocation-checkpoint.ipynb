```python
import numpy as np
import matplotlib.pyplot as plt
```

# Latent Dirichlet Allocation(潜在的ディレクレ分配法)

## トピックモデルの生成

```python
from gensim import corpora,models,similarities
```

```python
corpus = corpora.BleiCorpus("./AP-dataset/ap/ap.dat","./AP-dataset/ap/vocab.txt") # load corpus
```

```python
model = models.ldamodel.LdaModel(corpus,num_topics=100,id2word=corpus.id2word) # モデル生成
```

```python
topics = [model[c] for c in corpus] 
topics[0] # topics[0]のトピックをそのweightを表示
```

```python
plt.figure(facecolor="white")
plt.hist([len(t) for t in topics])
plt.xlabel("number of topic")
plt.ylabel("number of document")
plt.show()
```

```python
# 最初の10topicの10単語を表示
for t in range(10):
    for w in model.show_topic(t, 10):
        print(w[0],end=" ")
    print("")
```

## WordCloud

```python
import math
from wordcloud import WordCloud
```

```python
fig, axs = plt.subplots(ncols=2, nrows=math.ceil(5), figsize=(16,20))
axs = axs.flatten()

def color_func(word, font_size, position, orientation, random_state, font_path):
    return 'darkturquoise'

for i, t in enumerate(range(10)):

    x = dict(model.show_topic(t, 30))
    im = WordCloud(
        background_color='black',
        color_func=color_func,
        max_words=4000,
        width=300, height=300,
        random_state=0
    ).generate_from_frequencies(x)
    axs[i].imshow(im.recolor(colormap= 'Paired_r' , random_state=244), alpha=0.98)
    axs[i].axis('off')
    axs[i].set_title('Topic '+str(t))

# vis
plt.savefig('wordcloud.png') 
plt.show()
```
