```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## モデルの評価方法
　機械学習の目的は未知の予測をすることだから,未知データに対する予測性能を正しく評価する必要がある.このため,現在持っているデータセットを訓練(train)データと検証(test)データの2に分けて学習と評価を行う.モデルが1つのときは訓練用と検証用の2つにデータを分割すればよいが,多くの場合,複数のモデルを作成して最も性能の良いモデルを選ぶという手順がある.学習と検証を繰り返して得られるモデルは現在の評価データに対して最適な性能を示すモデルになる,すなわち,評価データもパラメータ選択の学習に使用しているということになる.このモデルは評価データに適合しているため,将来にとられる未知のデータに対して有効である保証がなくなる.  
　そこで数少ないデータセットを活用してモデルの汎化性能を正しく評価できるデータセット分割の方法として`ホールドアウト法`と`k-fold交差検証法`がある.

## ホールドアウト法
ホールドアウト法は全訓練データを訓練データ,検証データ,テストデータの3つに分ける方法である.
- 訓練データ : モデルのパラメータを推定するために使用されるデータ
- 検証データ : モデル構築時のハイパーパラメータをチューニングするために使用されるデータ
- テストデータ : 最適なパラメータとハイパーパラメータで構築されたモデルの性能を評価するためのデータ
ホールドアウト法では,まずモデルの学習を訓練データで行い,検証データで検証を行う.訓練データと検証データでハイパーパラメータの組み合わせの評価を行い,最後にテストデータで採取的な評価を行う.

## ホールドアウト法の例(Redge回帰)

```python
df = pd.read_csv("data3.csv")
df.head()
```

```python
plt.figure(facecolor='white')
plt.scatter(df["x"],df["y"],label='point')
plt.legend()
```

### データセットを訓練,検証,テストデータに分ける

```python
from sklearn.model_selection import train_test_split

X = df["x"].to_numpy().reshape(-1,1)
y = df["y"].to_numpy().reshape(-1,1)

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1) # 訓練データとテストデータに分割
X_train,X_valid,y_train,y_valid = train_test_split(X_train,y_train,random_state=1) # 訓練データと検証データに分割

print(X_train.shape) # 訓練データ
print(X_valid.shape) # 検証データ
print(X_test.shape) # テストデータ
```

### モデルの学習と検証データによるバリデーション

```python
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

degree = 4 # 次元は4で決め打ち
alphalist = [0,0.1,1,10,100]
Xpred = np.arange(-2,7,0.1).reshape(-1,1) # プロット用の行列を生成

poly_features = PolynomialFeatures(degree=degree,include_bias=False)
X_train_poly = poly_features.fit_transform(X_train)
X_valid_poly = poly_features.fit_transform(X_valid)

plt.figure(facecolor="white")
for alpha in alphalist:
    model = Ridge(alpha=alpha,solver="cholesky") # コレスキーの行列分解テクニック
    model.fit(X_train_poly,y_train)
    print("alpha : ",alpha)
    #print("切片 : ",model.intercept_[0])
    #print("傾き : ",model.coef_)
    print("決定係数(train) : ",model.score(X_train_poly,y_train)) # 決定係数を計算する説明変数X,目的変数y
    print("MSE(train) : ", mean_squared_error(y_train,model.predict(X_train_poly)))
    print("決定係数(valid) : ",model.score(X_valid_poly,y_valid)) # 決定係数を計算する説明変数X,目的変数y
    print("MSE(valid) : ", mean_squared_error(y_valid,model.predict(X_valid_poly)))
    print("") # 見た目
    Xpred_poly = poly_features.fit_transform(Xpred) # フォーマット変換
    ypred = model.predict(Xpred_poly) # 予測
    plt.plot(Xpred,ypred,label="alpha = "+str(alpha))

plt.scatter(df["x"],df["y"])
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.legend(fontsize=12)
plt.show()
```

ホールドアウト法の結果から,ハイパーパラメータ$\alpha=1$を採用する.

### テストデータによる最終評価

```python
X_test_poly = poly_features.fit_transform(X_test)
model = Ridge(alpha=1,solver="cholesky") # コレスキーの行列分解テクニック
model.fit(X_test_poly,y_test)
```

```python
Xpred = np.arange(-2,7,0.1).reshape(-1,1) # プロット用の行列を生成
Xpred_poly = poly_features.fit_transform(Xpred) # フォーマット変換
ypred = model.predict(Xpred_poly) # 予測

print("決定係数(test) : ",model.score(X_test_poly,y_test)) # 決定係数を計算する説明変数X,目的変数y
print("MSE(test) : ", mean_squared_error(y_test,model.predict(X_test_poly)))

plt.figure(facecolor="white")
plt.scatter(X_test,y_test)
plt.plot(Xpred,ypred,c="r",label="regression")
plt.xlabel("x",fontsize=15)
plt.ylabel("y",fontsize=15)
plt.legend()
plt.show()
```

```python
# 残差
y_pred = model.predict(X_test_poly)
resid = y_pred-y_test

plt.figure(facecolor="white")
plt.scatter(y_pred,resid)
plt.hlines(y = 0, xmin = 2, xmax = 12, color = 'black') # 要調整
plt.xlabel("予測値",fontsize=15) # predicted value
plt.ylabel("残差",fontsize=15) # residuals
plt.grid()
plt.show()
```

```python
from scipy import stats

# カーネル密度
plt.figure(facecolor='white')
sns.distplot(resid)
plt.show()

# 正規QQプロット
plt.figure(facecolor='white')
stats.probplot(resid.reshape(1,-1)[0],dist=stats.norm,plot=plt)
plt.xlabel("Theoretical quantiles",fontsize = 15)
plt.ylabel("Ordered Values",fontsize = 15)
plt.title("Probability Plot",fontsize = 17)
plt.show()
```

残差の中央は0であり,おおむね正規分布に従っているように見えるが,外れ値があることがわかる.

## k-fold交差検証法
ホールドアウト法の問題点として,モデルの分割方法が最終評価に大きく影響することがあげられる.すなわちホールドアウト法では汎化性能を評価できない場合が考えれれる.このようなケースを回避するためにk-fold検証を用いる.k-fold交差検証法では,データセットを訓練データとテストデータに分割した後に,訓練データをk個のサブセットに分割して,k-1個のサブセットで学習を行い,残りの1つのサブセットで検証を行う.例えば,訓練データをS1からS5の5つに分割したとする.そして,S1を検証データ,S2からS5を訓練データとして学習と検証を行い,性能指標を計算する.次にS2を検証データ,残りのデータを訓練データとして学習と検証を行い,性能指標を計算する.これをすべてのサブセットが検証データになるまで繰り返す.得られた性能指標の平均値を計算し,これを最終的な評価指標とする.

## k-分割交差検証法の例(Redge回帰)

### データセットを分割する

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.model_selection import cross_val_score # 交差検証のためにデータを分割するクラス

X = df["x"].to_numpy().reshape(-1,1)
y = df["y"].to_numpy().reshape(-1,1)

degree=4
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1) # 訓練データとテストデータに分割
poly_features = PolynomialFeatures(degree=degree,include_bias=False)
X_train_poly = poly_features.fit_transform(X_train)

model = Ridge(alpha=1,solver="cholesky")

scores = cross_val_score(model, X_train_poly, y_train, cv=5,scoring="r2")
# n_jobsの設定可能
# r2 決定係数で評価
# neg_mean_squared_error : MSEで評価
print(scores)
print(scores.mean())
```
