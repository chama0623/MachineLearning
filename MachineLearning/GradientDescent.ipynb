# Gradient Descent(勾配降下法)

勾配降下法で$z=x^2+y^2$が最小となるような$(x,y)$を求める. 理論値は$(x,y)=(0,0)$のときzが最小で$z=0$.

```python
def gradient_descent(x0,y0,eta=0.1,max_iter=1000):
    """勾配降下法の計算を行う関数
    
    Args:
    x0 : xの初期値
    y0 : yの初期値
    eta : learning rate
    max_iter : 反復回数
    """
    
    # 計算結果を格納
    xpred = [x0]
    ypred = [y0]
    vec=[]
    
    x=x0
    y=y0
    # 勾配降下法
    for i in range(max_iter):
        x,y = np.array([x,y]) - eta*df(x,y)
        print(i+1,x,y)
        if (abs(xpred[-1]-x)<=10e-6): # 収束判定
            print("converge")
        xpred.append(x)
        ypred.append(y)   
    
    # zを計算 
    xpred = np.array(xpred)
    ypred = np.array(ypred)
    #zpred = f(xpred,ypred)    
    x = np.arange(-10,11,2)
    y = np.arange(-10,11,2)
    #z = f(x,y)
    
    # メッシュを生成
    X,Y = np.meshgrid(x,y)
    Z = f(X,Y)
    
    # figure生成
    fig = plt.figure()
    #ax1 = Axes3D(fig)
    
    #ax1.set_xlabel("x")
    #ax1.set_ylabel("y")
    #ax1.set_zlabel("z")

    #ax1.plot_wireframe(X,Y,Z,label="f(x,y)")
    #ax1.scatter3D(xpred,ypred,zpred,label="gb",color="red",s=50)
    cont = plt.contour(X, Y, Z,levels=range(5,200,15))
    cont.clabel(fmt='%1.1f', fontsize=14)
    plt.gca().set_aspect('equal')
    plt.scatter(xpred,ypred,color="r")
    plt.plot(xpred,ypred,color="r")
    plt.show()
```

```python
# 最小化する関数を定義
def f(x,y):
    return x**2 + y**2

# 最小化する関数の勾配ベクトルを定義
def df(x,y):
    return np.array([2*x,2*y])

eta = 0.1 # learning rate
max_iter = 1000
x0 = -10
y0 = 10

gradient_descent(x0,y0,max_iter=20)
```
