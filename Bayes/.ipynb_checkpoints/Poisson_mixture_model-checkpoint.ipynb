```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, gamma
```

# ポアソン混合分布

まず, 観測モデルを設定する. K個の観測モデルをポアソン分布Po($x_i | \lambda_k$)とする. そして各データに対するK個の観測モデルの割り当てをカテゴリ分布Cat($s_i | \pi$)で行う.

```python
# 真のlambdaパラメータ(K=3)
lambda_truth_k = np.array([10.0,25.0,40.0])

# 真の混合比率
pi_truth_k = np.array([0.35,0.25,0.4])

# クラスタ数
K = len(lambda_truth_k)
```

```python
# ポアソン混合分布の確率密度関数を描画
x_line = np.arange(0,2*np.max(lambda_truth_k)) # 0からlambdaパラメータの最大値の2倍までを描画

# 確率分布関数を計算
model_prob = 0.0 
for k in range(K):
    # k番目の確率分布関数
    tmp_prob = poisson.pmf(x_line,mu=lambda_truth_k[k])
    # 混合比率で重みづけた加重平均
    model_prob += pi_truth_k[k]*tmp_prob
    
plt.figure(facecolor="white")
plt.bar(x_line,model_prob)
plt.xlabel("x")
plt.ylabel("prob")
plt.title("Poisson mixture model")
plt.show()
```

データを生成する. 潜在変数$S={s_1, s_2, \cdots, s_n}$, $s_i=(s_{i,1},s_{i,2},\cdots,s_{i,n})$を生成する. そして各クラスに割り当てられたクラスタに従って観測データ$X={x_1,x_2, \cdots, x_n}$を生成する.

```python
# データ数
n = 250

# 潜在変数を生成
s_truth_nk = np.random.multinomial(n=1,pvals=pi_truth_k,size=n)
print("潜在変数Sの先頭5列")
print(s_truth_nk[:5])
print("潜在変数の行列のサイズ")
print(s_truth_nk.shape)

# クラスタ番号を抽出
_,s_truth_n = np.where(s_truth_nk==1)
print("\n属するクラスタ")
print(s_truth_n)

# 観測データを生成
x_i = np.random.poisson(lam=lambda_truth_k[s_truth_n],size=n)
print("\n観測データの先頭10行")
print(x_i[:10])

# 観測データと確率分布関数のプロット
plt.figure(facecolor="white")
plt.bar(x_line,model_prob,color='white', alpha=1, edgecolor='red', linestyle='--',label="確率分布関数") # 確率分布関数
plt.bar(x=x_line,height=[np.sum(x_i==x)/len(x_i) for x in x_line],label="観測データ") # 観測データ
plt.xlabel("x")
plt.ylabel("density")
plt.title("Poisson mixture model")
plt.legend()
plt.show()

# クラスタごとに色を分けて描画
plt.figure(facecolor="white")
plt.bar(x_line,model_prob,color='white', alpha=1, edgecolor='red', linestyle='--',label="確率分布関数") # 確率分布関数
for k in range(K):
    plt.bar(x=x_line,height=[np.sum(x_i[s_truth_n==k]==x)/len(x_i) for x in x_line],label="クラスタ"+str(k+1),alpha=0.5)# # 観測データ
plt.xlabel("x")
plt.ylabel("density")
plt.title("Poisson mixture model")
plt.legend()
plt.show()
```

# Gibbs sampling

　ポアソン混合モデルでGibbs samplingを行う. まず, 観測モデルの分布(ポアソン分布)$p(X|\lambda)$と潜在変数の分布(カテゴリ分布)$p(S|\pi)$に対して共役事前分布を設定する. ポアソン分布のパラメータ$\lambda$(K次元ベクトル)の事前分布は$Ga(\lambda | a,b)$, カテゴリ分布の事前分布としてディリクレ分布$p(\pi | \alpha)$を設定する. ガンマ分布のパラメータ$b$は「現代数理統計学の基礎」に定義されている$b$として扱う. このときにガンマ分布をプロットしておく.  
　次に事前分布に従ってランダムに生成した値でパラメータ$\lambda,\pi$を初期化する. このときに初期値で生成した値でポアソン混合分布を生成してプロットしておく. そして次に示す手順でギブスサンプリングを行う.
 1. 反復回数n_iter回次の処理を行う.
     1. 潜在変数の事後分布を計算する.
     2. 事後分布から潜在変数をn回サンプリングする.
     3. $\lambda$の事後分布を計算する.
     4. 事後分布から$\lambda$をK回サンプリングする.
     5. $\pi$の事後分布を計算する.
     6. 事後分布から$\pi$をサンプリングする.

```python
# lambdaの事前分布
a=1.0
b=1.0

# piの事前分布
alpha_k = np.repeat(2.0,K)

# 事前分布をプロット
lambda_line = np.linspace(0.0, 2.0 * np.max(lambda_truth_k), num=1000)
prior_lambda = gamma.pdf(x=lambda_line, a=a, scale=1 / b) # lambdaの密度関数

plt.figure(facecolor="white")
plt.plot(lambda_line,prior_lambda,label="prior distribution") # 事前分布
plt.vlines(x=lambda_truth_k,ymin=0.0,ymax=np.max(prior_lambda),color='red', linestyle='--',label="true value") # 真の分布
plt.xlabel("x")
plt.ylabel("density")
plt.title("Poisson mixture model")
plt.legend()
plt.show()
```

```python
# 初期値の設定
lambda_k = np.random.gamma(shape=a,scale=1/b,size=K)
pi_k = np.random.dirichlet(alpha=alpha_k,size=1).reshape(K)
print("lambda init : ",lambda_k)
print("pi init : ",pi_k)

# 初期値から生成したポアソン混合分布の分布関数をプロット
# 確率分布関数を計算
init_prob = 0.0 
for k in range(K):
    # k番目の確率分布関数
    tmp_prob = poisson.pmf(x_line,mu=lambda_k[k])
    # 混合比率で重みづけた加重平均
    init_prob += pi_k[k]*tmp_prob
    
plt.figure(facecolor="white")
plt.bar(x_line,init_prob, edgecolor='red', linestyle='--', label="初期値から生成した分布") # 初期値から生成した分布
plt.bar(x_line,model_prob,label="真の分布") # 真の分布
plt.xlabel("x")
plt.ylabel("prob")
plt.title("Poisson mixture model")
plt.show()
```

```python
# ギブスサンプリング
n_iter = 100

eta_nk = np.empty((n,K))
s_nk = np.empty((n,K))

trace_s_in = [[np.nan] * n]
trace_a_ik = [[a] * n]
trace_b_ik = [[b] * n]
trace_alpha_ik = [list(alpha_k)]
trace_lambda_ik = [list(lambda_k)]
trace_pi_ik = [list(pi_k)]

for i in range(n_iter):
    for j in range(n):
        # 潜在変数の事後分布を計算
        tmp_eta_k = np.exp(x_i[j] * np.log(lambda_k) - lambda_k + np.log(pi_k))
        eta_nk[j] = tmp_eta_k / np.sum(tmp_eta_k) # 正規化
        # クラスタをサンプリング
        s_nk[j] = np.random.multinomial(n=1, pvals=eta_nk[j])
        
    # lambdaの事後分布を計算
    a_hat_k = np.sum(s_nk.T * x_i, axis=1) + a
    b_hat_k = np.sum(s_nk, axis=0) + b
    # lambdaをサンプリング
    lambda_k = np.random.gamma(shape=a_hat_k, scale=1 / b_hat_k, size=K)
        
    # piの事後分布を計算
    alpha_hat_k = np.sum(s_nk, axis=0) + alpha_k
    # piをサンプリング
    pi_k = np.random.dirichlet(alpha=alpha_hat_k, size=1).reshape(K)
        
    # 値を記録
    _, s_n = np.where(s_nk == 1) # クラスタ番号を抽出
    trace_s_in.append(list(s_n))
    trace_a_ik.append(list(a_hat_k))
    trace_b_ik.append(list(b_hat_k))
    trace_alpha_ik.append(list(alpha_hat_k))
    trace_lambda_ik.append(list(lambda_k))
    trace_pi_ik.append(list(pi_k))
```

```python
# 最後のa,bを用いて事後分布を計算
posterior_lambda_k = np.empty((K, len(lambda_line)))
for k in range(K):
    posterior_lambda_k[k] = gamma.pdf(x=lambda_line, a=a_hat_k[k], scale=1 / b_hat_k[k])
    
# 事後分布をプロット
plt.figure(facecolor="white")
for k in range(K):
    plt.plot(lambda_line,posterior_lambda_k[k],label="クラスタ"+str(k+1))
plt.vlines(x=lambda_truth_k,ymin=0.0,ymax=np.max(prior_lambda),color='red', linestyle='--',label="true value") # 真の分布
plt.xlabel("x")
plt.ylabel("density")
plt.title("Poisson mixture model")
plt.legend()
plt.show()
```

```python
# 事後分布から生成した混合分布の確率分布関数をプロット
pro_prob = 0.0 
for k in range(K):
    # k番目の確率分布関数
    tmp_prob = poisson.pmf(x_line,mu=lambda_k[k])
    # 混合比率で重みづけた加重平均
    pro_prob += pi_k[k]*tmp_prob
    
plt.figure(facecolor="white")
plt.bar(x_line,pro_prob, edgecolor='red', linestyle='--', label="初期値から生成した分布") # 事後分布から生成した分布
plt.bar(x_line,model_prob,label="真の分布") # 真の分布
plt.xlabel("x")
plt.ylabel("prob")
plt.title("Poisson mixture model")
plt.show()
```
