```python
import numpy as np
import matplotlib.pyplot as plt
```

# ベルヌーイ分布のエントロピー

ベルヌーイ分布のエントロピー$H$は次で表される.

$$H = -p \log p -(1-p)\log(1-p)$$

```python
# ベルヌーイ分布のエントロピーの図示
def bernoulli_entropy(p):
    return -p*np.log(p)-(1-p)*np.log(1-p)

p = np.arange(0.01,1,0.01)
H = bernoulli_entropy(p)

plt.figure(facecolor="white")
plt.plot(p,H)
plt.xlim(0,1)
plt.title("ベルヌーイ分布のエントロピー")
plt.show()
```

ベルヌーイ分布のエントロピーは0および1のとき最小, 0.5のときに最大になる. エントロピーは乱雑さの度合いを表すから0および1のときは成功または失敗が決まった事象であるからエントロピーは最小になる. 0.5のときは成功または失敗が等確率であるからエントロピーが最大になる.
