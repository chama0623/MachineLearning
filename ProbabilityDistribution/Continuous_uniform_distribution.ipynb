# Continuous Uniform Distribution(連続一様分布) : 連続

## 確率密度関数
$$
f(x) = 
\begin{cases}
\frac{1}{b-a} (a \leq x \leq b)& \\
0 (otherwise)& \\
\end{cases}
$$

モーメント母関数
$$
M_X(t) = \frac{e^{tb}-e^{ta}}{t(b-a)}
$$

期待値$$E[X] = \frac{a+b}{2}$$  分散$$V[X] = \frac{(b-a)^2}{12}$$

共役事前分布 : なし  

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
a,b = 1,3

xscale = np.linspace(-5,5,100)

prob = stats.uniform.pdf(xscale,loc=a, scale=b-a)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

## 累積分布関数のグラフ

```python
a,b = 1,4
xscale = np.linspace(0,5,100)

prob = stats.uniform.pdf(xscale,a,b-a)
cdf = stats.uniform.cdf(xscale,a,b-a)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```
