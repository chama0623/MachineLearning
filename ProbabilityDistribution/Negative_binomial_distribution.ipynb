## Negative binomial Distribution(負の二項分布) : 離散


確率pで成功する試行をr回成功するまでの失敗回数Xが従う分布

確率密度関数

$$
f(x)={}_{r+x-1}C_x p^r(1-p)^{x}
$$

積率母関数 : $$ \frac{p^r}{(1-s(1-p))^r} $$

期待値$$E[X] = r\frac{1-p}{p}$$ 分散$$V[X] = r\frac{1-p}{p^2}$$

他の分布との関係
- r=1のとき幾何分布Geo(p)である.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
k,p=5,0.5
xscale = np.linspace(0,20,20+1)

prob = stats.nbinom.pmf(xscale,k,p) 

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("NBi("+str(k)+","+str(p)+")",fontsize=15)
plt.show()
```

```python
plist=[0.5,0.5,0.5]
klist=[5,6,7]

xscalelist = []
problist = []
mulist = []
vlist = []
for i in range(len(plist)):
    xscalelist.append(np.linspace(0,30,30+1))
    problist.append(stats.nbinom.pmf(xscalelist[i],klist[i],plist[i]))
    mulist.append(klist[i]*(1-plist[i])/plist[i])
    vlist.append(klist[i]*(1-plist[i])/plist[i]**2)

plt.figure(facecolor="white")
for i in range(len(plist)):
    plt.plot(xscalelist[i],problist[i],label="NBi("+str(plist[i])+")")
    plt.scatter(xscalelist[i],problist[i])
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

## 累積分布関数のグラフ

```python
k,p = 5,0.5
xscale = np.linspace(0,20,20+1)

prob = stats.nbinom.pmf(xscale,k,p)
prob_cd = stats.nbinom.cdf(xscale,k,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("NBi("+str(k)+","+str(p)+")",fontsize=15)
plt.legend()
plt.show()
```
