# Exponential distribution(指数分布) : 連続

## 確率密度関数
$$
f(X)=\lambda e^{-\lambda x}
$$

モーメント母関数
ガンマ分布Ga(1,$\frac{1}{\lambda}$)に等しい
$$
M_X(t) = \frac{\lambda}{\lambda - t }
$$

期待値$$E[X] = \frac{1}{\lambda}$$ 分散$$V[X] = \frac{1}{\lambda^2}$$


再生性 : なし

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
l = 1
xscale = np.linspace(0,5,120)

prob = stats.expon.pdf(xscale,1/l)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

ドキュメントの次の文章に注意する.
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html#scipy.stats.expon  
A common parameterization for expon is in terms of the rate parameter lambda, such that pdf = lambda * exp(-lambda * x). This parameterization corresponds to using scale = 1 / lambda.  
だいたいの訳 : パラメータは1/lambdaで指定.

```python
llist=[0.5,1,1.5]

xscalelist = []
problist = []
mulist = []
vlist = []
for i in range(len(llist)):
    xscalelist.append(np.linspace(0,10,200))
    problist.append(stats.expon.pdf(xscalelist[i],1/llist[i]))
    mulist.append(1/llist[i])
    vlist.append(1/llist[i]**2)
    
plt.figure(facecolor="white")
for i in range(len(llist)):
    plt.plot(xscalelist[i],problist[i],label="E("+str(llist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.ylabel("P(X)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

## 累積分布関数のグラフ

```python
l = 1
xscale = np.linspace(0,5,120)

prob = stats.expon.pdf(xscale,1/l)
cdf = stats.expon.cdf(xscale,1/l)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```
