# Chi-square Distribution($\chi^2$分布) : 連続

## 確率密度関数
自由度nとする.
$$
f(X)=\frac{x^{\frac{n}{2}-1} e^{-\frac{x}{2}}}{2^{\frac{n}{2}}\Gamma \left(\frac{n}{2} \right)}
$$

モーメント母関数
$$
M_X(t) = \left( \frac{1}{1-2t} \right)^{\frac{n}{2}}
$$

期待値$$E[X] = n$$ 分散$$V[X] = 2n$$

共役事前分布 : なし  
再生性 : あり

## 確率分布関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n=3 # degree of freedom

xscale = np.linspace(0,10,100)

prob = stats.chi2.pdf(xscale,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

```python
nlist=[1,2,3,4,5] # degree of freedom
width=[0,8] # 表示幅調整

xscalelist = []
problist = []
vlist = []
skewlist = []
kurtosislist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(width[0],width[1],100))
    problist.append(stats.chi2.pdf(xscalelist[i],nlist[i]))
    vlist.append(2*nlist[i])
    skewlist.append(stats.skew(problist[i]))
    kurtosislist.append(stats.kurtosis(problist[i]))
    
plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="カイ2乗("+str(nlist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.ylabel("P(X)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(nlist))
print("Variance : "+str(vlist))
print("Skewness : "+str(skewlist)) #歪度
print("Kurtosislist : "+str(kurtosislist)) #尖度
```

## 累積分布関数のグラフ

```python
n=4
xscale = np.linspace(0,10,100)

prob = stats.chi2.pdf(xscale,n)
cdf = stats.chi2.cdf(xscale,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```
