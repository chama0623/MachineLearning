# Hypergeometric Distribution(超幾何分布) : 離散

## 確率密度関数
2種類のA,BからなるN個のものがあり,個数の構成はそれぞれM,M-N個とする.この集団からn個を非復元抽出したとき
Aがx個,Bがn-x個であるとするとx=kである確率は超幾何分布に従う.またその確率密度関数は
$$
P(X=k) = {}_M \mathrm{C} _k \frac{{}_{N-M} \mathrm{C} _{n-k}}{{}_N \mathrm{C} _n}
$$

モーメント母関数 : 一般に存在しない.

期待値$$E[X] = n\frac{M}{N}$$ 分散$$V[X] = n\frac{M(N-M)(N-n)}{N^2(N-1)}$$

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
N,M,n = 1000,200,5
xscale = np.linspace(0,n,n+1)

prob = stats.hypergeom.pmf(xscale,N,M,n) # N,M,nの順番はドキュメント参照

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("H("+str(N)+","+str(M)+","+str(n)+")",fontsize=15)
plt.show()
```

```python
Nlist=[1000,2000]
Mlist=[200,700]
nlist = [5,8]

xscalelist = []
problist = []
mulist = []
vlist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(0,nlist[i],nlist[i]+1))
    problist.append(stats.hypergeom.pmf(xscalelist[i],Nlist[i],Mlist[i],nlist[i]))
    mulist.append(nlist[i]*Nlist[i]/Mlist[i])
    vlist.append(nlist[i]*Mlist[i]*(Nlist[i]-Mlist[i])*(Nlist[i]-nlist[i])/(Nlist[i]**2 *(Nlist[i]-1)))

plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="H("+str(Nlist[i])+","+str(Mlist[i])+","+str(nlist[i])+")")
    plt.scatter(xscalelist[i],problist[i])
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

## 累積分布関数のグラフ

```python
N,M,n = 1000,200,5
xscale = np.linspace(0,n,n+1)

prob = stats.hypergeom.pmf(xscale,N,M,n)
prob_cd = stats.hypergeom.cdf(xscale,N,M,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("H("+str(N)+","+str(M)+","+str(n)+")",fontsize=15)
plt.legend()
plt.show()
```
