## Geometric Distribution(幾何分布) : 離散

確率pで成功する事象が初めて成功するまでの失敗回数Xが従う分布

確率密度関数
$$
f(x) = p(1-p)^{x} 
$$

確率母関数 :$$ G_X(s) = \frac{p}{1-s(1-p)}$$
ただし $$s<\frac{1}{1-p}$$

期待値$$E[X] = \frac{1-p}{p}$$ 分散$$V[X] = \frac{1-p}{p^2}$$
特徴 : 無記憶性  
他の分布との関係 : 
- 二項分布において,初めて成功するまでの失敗回数が従う分布である.
- 初めてr回成功するまでの失敗回数Xは負の二項分布NB(r,p)に従う.幾何分布はNB(1,p)である.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
p = 1/6
xscale = np.linspace(1,10,10)

prob = stats.geom.pmf(xscale,p) 

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Geo("+str(p)+")",fontsize=15)
plt.show()
```

```python
plist=[0.2,0.5,0.8]

xscalelist = []
problist = []
mulist = []
vlist = []
for i in range(len(plist)):
    xscalelist.append(np.linspace(1,10,10))
    problist.append(stats.geom.pmf(xscalelist[i],plist[i]))
    mulist.append(1/plist[i])
    vlist.append((1-plist[i])/plist[i]**2)

plt.figure(facecolor="white")
for i in range(len(plist)):
    plt.plot(xscalelist[i],problist[i],label="G("+str(plist[i])+")")
    plt.scatter(xscalelist[i],problist[i])
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

## 累積分布関数のグラフ

```python
p = 0.04
xscale = np.linspace(1,10,10)

prob = stats.geom.pmf(xscale,p)
prob_cd = stats.geom.cdf(xscale,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("G("+str(p)+")",fontsize=15)
plt.legend()
plt.show()
```
