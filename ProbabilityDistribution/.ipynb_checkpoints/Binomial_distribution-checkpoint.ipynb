# Binomial Distribution(二項分布) : 離散

## 確率密度関数
nを試行回数,pを成功率とすると確率密度関数は
$$
P(X=k)={}_nC_kp^k(1-k)^{n-k}
$$

確率母関数
$$
G_X(s) = (sp+1-p)^n
$$

期待値$$E[X] = np$$ 分散$$V[X] = np(1-p)$$

共役事前分布 : Beta分布  
再生性 : あり  
使用例 : コインをn回投げたとき,表が出る回数XはBi(n,1/2)に従う.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n,p = 10,0.5
xscale = np.linspace(0,n,n+1)

prob = stats.binom.pmf(xscale,n,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Bi("+str(n)+","+str(p)+")",fontsize=15)
plt.show()
```

```python
nlist=[20,20,20]
plist=[0.2,0.5,0.8]

xscalelist = []
problist = []
mulist = []
vlist = []
skewlist = []
kurtosislist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(0,nlist[i],nlist[i]+1))
    problist.append(stats.binom.pmf(xscalelist[i],nlist[i],plist[i]))
    mulist.append(nlist[i]*plist[i])
    vlist.append(nlist[i]*plist[i]*(1-plist[i]))
    skewlist.append(stats.skew(problist[i]))
    kurtosislist.append(stats.kurtosis(problist[i]))

plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="Bi("+str(nlist[i])+","+str(plist[i])+")")
    plt.scatter(xscalelist[i],problist[i])
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
print("Skewness : "+str(skewlist)) #歪度
print("Kurtosislist : "+str(kurtosislist)) #尖度
```

## 累積分布関数のグラフ

```python
n,p = 10,0.5
xscale = np.linspace(0,n,n+1)

prob = stats.binom.pmf(xscale,n,p)
prob_cd = stats.binom.cdf(xscale,n,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("試行回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Bi("+str(n)+","+str(p)+")",fontsize=15)
plt.legend()
plt.show()
```

## 二項分布による超幾何分布の近似

```python
N,M,Hn = 300,40,5
Bn,p = Hn,M/N
xscale = np.linspace(0,Hn,Hn+1)

probH = stats.hypergeom.pmf(xscale,N,M,Hn) # N,M,nの順番はドキュメント参照
probB = stats.binom.pmf(xscale,Bn,p)

plt.figure(facecolor="white")
plt.plot(xscale,probH,label="H("+str(N)+","+str(M)+","+str(Hn)+")")
plt.scatter(xscale,probH)
plt.plot(xscale,probB,label="Bi("+str(Bn)+","+str(p)+")")
plt.scatter(xscale,probB)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
```
