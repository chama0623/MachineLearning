```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## MNISTデータ
MNISTデータは70000件の手書き文字画像のデータで,全ての画像が28x28ピクセル,0から255のグレースケールで構成されている.

```python
# load data
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1,)
X,y = mnist["data"],mnist["target"]
print(X.shape)
print(y.shape)
```

```python
%matplotlib inline
import matplotlib

# 適当な番号のデータを1つ抽出
some_digit = X[36000]
# 28x28にreshape
some_digit_img = some_digit.reshape(28,28)

# cmap : 色の指定
# matplotlib.cm.binary : バイナリヒートマップ
# interpolation : nearestにすると小さい画像を拡大表示するときに補完してくれる
# plt.axis() : offにすると軸を非表示にする
plt.imshow(some_digit_img,cmap = matplotlib.cm.binary,interpolation="nearest")
plt.axis("off")
plt.show()
```

## PCA

```python
# PCA実行
from sklearn.decomposition import PCA

# n_compomemts : 射影する次元
pca = PCA(n_components=10) # 10次元
X_red = pca.fit_transform(X)

# PC1,PC2をプロット
plt.figure(facecolor="white")
plt.scatter(X_red[:,0],X_red[:,1])
plt.title("decomposed data")
plt.show()
```

```python
# 因子寄与率
pca.explained_variance_ratio_
```

```python
# 適切な次元の選択
cri = [0.8,0.85,0.9,0.95] # 累積寄与率基準
pca = PCA(n_components=28*28)
X_red = pca.fit_transform(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
for cr in cri:
    d = np.argmax(cumsum>=cr)+1
    print("累積寄与率基準 : %.3lf best dimension : %d" %(cr,d))
    
plt.figure(facecolor="white")
plt.plot(range(1,cumsum.shape[0]+1),cumsum)
plt.grid()
plt.xlabel("Dimention",fontsize=15)
plt.ylabel("Cumulative contribution rate",fontsize=15)
plt.show()
```

累積寄与率を見ると第一主成分の分散は全体の9.7%,第二主成分の分散は全体の7.2%であることがわかる. オリジナルデータの次元が784と大きいため特別分散の大きい主成分が得られていないことがわかる. 43次元のとき累積寄与率基準が0.8となり,154次元のとき累積寄与率基準ｇな0.9を満たす. オリジナルデータの次元数が784であるから十分次元削減出来ていると考えられる.

```python
# n_componentsに0から1未満の少数を与えると,与えた小数の累積寄与率基準でPCAを実行する
pca = PCA(n_components=0.95)
X_red = pca.fit_transform(X)
print(X)
```

## PCA逆変換

```python
pca = PCA(n_components=154) # 累積寄与率基準0.95でPCAを実行
X_red = pca.fit_transform(X)
X_rec = pca.inverse_transform(X_red) # 逆変換

# 再構築誤差
mse = (X-X_rec).mean(axis=1)
print(mse[:5])
print("mean : ",mse.mean())
print("std : ",mse.std())
```

## 追加学習型PCA
SVDアルゴリズムは訓練セット全体がメモリに収まっている必要があり,メモリ管理の問題が生じる. 追加学習型PCA(IPCA)は訓練セットをミニバッチに分解してPCAを実行する. このように新しいインスタンスに対しても実行できることをオンライン実行と呼ぶ.

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100 # 訓練セットを100分割
inc_pca = IncrementalPCA(n_components=154)
# 訓練セットを分割してIPCAを実行
for X_batch in np.array_split(X,n_batches):
    inc_pca.partial_fit(X_batch)
    
X_rec = inc_pca.transform(X)
pca.explained_variance_ratio_[:10]
```

## ランダム化PCA
PCA実行のためのもう一つのオプションとしてランダム化PCAがある. ランダム化PCAは最初のd個の主成分の近似値を高速で見つける確率的なアルゴリズムである.

```python
%%time
# svd_solverを引数で与えることでランダム化PCAを実行する
rnd_pca = PCA(n_components=154,svd_solver="randomized")
X_red = rnd_pca.fit_transform(X)
print(rnd_pca.explained_variance_ratio_[:5])
```

```python
%%time
# 実行時間比較用 通常のPCA
pca = PCA(n_components=154)
X_red = pca.fit_transform(X)
print(rnd_pca.explained_variance_ratio_[:5])
```

## Swiss rollデータ
Swiss rollデータは2次元多様体を生成するデータセットである. 2次元多様体とは高次元空間でねじったり回転したりすることができる2次元図形のことである.

```python
from sklearn.datasets import make_swiss_roll
X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
```

```python
axes = [-11.5, 14, -2, 23, -12, 15]

fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.hot)
ax.view_init(10, -70)
ax.set_xlabel("$x_1$", fontsize=18)
ax.set_ylabel("$x_2$", fontsize=18)
ax.set_zlabel("$x_3$", fontsize=18)
ax.set_xlim(axes[0:2])
ax.set_ylim(axes[2:4])
ax.set_zlim(axes[4:6])
plt.show()
```

## カーネル化PCA
SVM同様にカーネルトリックを用いることで,暗黙のうちにインスタンスを非常に高次の特徴量空間にマッピングして非線形射影をPCAに適用することができる.

Documents : https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html?highlight=kernelpca#sklearn.decomposition.KernelPCA

### RBFカーネル

```python
from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components=2,kernel="rbf",gamma=0.04)
X_red = rbf_pca.fit_transform(X)

plt.figure(facecolor="white")
plt.scatter(X_red[:,0],X_red[:,1],c=t, cmap=plt.cm.hot)
plt.xlabel("$z_1$",fontsize=15)
plt.ylabel("$z_2$",fontsize=15)
plt.show()
```

### シグモイドカーネル

```python
from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components=2,kernel="sigmoid",gamma=0.001)
X_red = rbf_pca.fit_transform(X)

plt.figure(facecolor="white")
plt.scatter(X_red[:,0],X_red[:,1],c=t, cmap=plt.cm.hot)
plt.xlabel("$z_1$",fontsize=15)
plt.ylabel("$z_2$",fontsize=15)
plt.show()
```

### 線形カーネル

```python
from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components=2,kernel="linear")
X_red = rbf_pca.fit_transform(X)

plt.figure(facecolor="white")
plt.scatter(X_red[:,0],X_red[:,1],c=t, cmap=plt.cm.hot)
plt.xlabel("$z_1$",fontsize=15)
plt.ylabel("$z_2$",fontsize=15)
plt.show()
```

## カーネルの選択とハイパーパラメータチューニング
KernelPCAは教師なし学習の手法であるから最良のカーネル,ハイパーパラメータを選択するための性能指標がない. しかし次元削減は教師あり学習タスクの準備ステップになっていることが多いためグリッドサーチを用いて最高の性能を引き出せるカーネルとハイパーパラメータを選択すればよい. ここではKernelPCAとロジスティック回帰のパイプラインを作成しカーネルの選択とハイパーパラメータのチューニングを行う.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

y = t > 6.9 # tを基準6.9で二値化

clf = Pipeline([
    ("kpca",KernelPCA(n_components=2)),
    ("log_reg",LogisticRegression())
])

params_grid = [{
    "kpca__gamma" : np.linspace(0.03,0.05,10),
    "kpca__kernel" : ["rbf","sigmoid"]
}]

grid_search = GridSearchCV(clf,params_grid,cv=3)
grid_search.fit(X,y)
print(grid_search.best_params_)
```
