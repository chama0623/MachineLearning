```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

## 分析の概要と目的
data2はあるコーヒーショップの各月のアイスコーヒーの売り上げ`Sales`と来店客数`Passengers`のデータである.来店客数を独立変数,アイスコーヒーの売り上げを従属変数として,単回帰分析によって来店客数からアイスコーヒーの売り上げを予測できるか検討することが目的である.

```python
# read data
df = pd.read_csv("data2.csv")
df
```

```python
# plot data
plt.figure(facecolor="white")
plt.scatter(df["Passengers"],df["Sales"])
plt.xlabel("来店客数(人)",fontsize=15)
plt.ylabel("アイスコーヒーの売り上げ(杯)",fontsize=15)
plt.show()
```

## scikit-learnを用いた回帰分析

```python
from sklearn.linear_model import LinearRegression

X = df["Passengers"].to_numpy().reshape(-1,1)
y = df["Sales"].to_numpy().reshape(-1,1)

model = LinearRegression()
model.fit(X,y)
print('回帰方程式 y = %.2fx + %.2f' % (model.coef_ , model.intercept_))
print("決定係数 %lf"%(model.score(X,y)))
```

```python
# plot data
plt.figure(facecolor="white")
plt.scatter(df["Passengers"],df["Sales"])
plt.plot(X,model.predict(X),c="red")
plt.xlabel("来店客数(人)",fontsize=15)
plt.ylabel("アイスコーヒーの売り上げ(杯)",fontsize=15)
plt.show()
```

## statsmodelsを用いた回帰分析

```python
import statsmodels.api as sm
model = sm.OLS(y, sm.add_constant(X))
results=model.fit()
print(results.summary())
```

## 結果の解釈
- 回帰式はy = 0.52x + -59.77である.
- t検定の結果,P値は0.00418で有意である.
- 自由度調整済み決定係数は0.534であり,このモデルでデータの53.4%を説明できる.
