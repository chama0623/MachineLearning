## 順序統計量(order statictics)の例

$X_1, \cdots X_n , i.i.d. ～ U(0,1)$のとき, $X_1,X_2, \cdots$を昇順に並べた$X_{(1)},X_{(2)},\cdots,X_{(n)}$を順序統計量という. $X_{(j)}$の確率密度関数は
$$f_{X_{(j)}}(x) ～ Beta(j,n-j-1)$$

これから$U(0,1)$から生成した乱数20個×1000セットについて最大値,最小値の分布を求め, 順序統計量の理論と比較してみる.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# シードを固定
np.random.seed(0)

n=20

# 乱数を生成
rand_set = np.random.random_sample([1000,n])
rand_set[0]
```

```python
# 最大値,最小値を計算
mins = np.min(rand_set,axis=1)
maxs = np.max(rand_set,axis=1)

# 確率密度関数を計算
xscale = np.linspace(0,1,100)
prob_min = stats.beta.pdf(xscale,1,n-1+1)
prob_max = stats.beta.pdf(xscale,n,n-n+1)
```

```python
plt.figure()
plt.hist(mins,density=True)
plt.plot(xscale,prob_min)
plt.title("最小値の分布")
plt.xlabel("最小値")
plt.ylabel("Density")
```

```python
plt.figure()
plt.hist(maxs,density=True)
plt.plot(xscale,prob_max)
plt.title("最大値の分布")
plt.xlabel("最大値")
plt.ylabel("Density")
```
