```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# 一様乱数生成

```python
nbins = 50
np.random.seed(0)
N = 100000

# 一様乱数生成
U = stats.uniform(loc=0.0,scale=1.0).rvs(size=N)
plt.figure(facecolor="white")
plt.hist(U,nbins, density=True)
plt.show()
```

# 逆関数法による乱数生成

## 指数分布

```python
# 1/lambda
scale=1.0

# 逆関数法
x1 = -scale*np.log(1-U)

# 理想的なpdf
rv= stats.expon(scale=scale)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
x2 = rv.pdf(x)

plt.figure(facecolor="white")
# 逆関数法で生成した乱数
plt.hist(x1,nbins, density=True)
# 理想的なpdf
plt.plot(x,x2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))
plt.show()
```
