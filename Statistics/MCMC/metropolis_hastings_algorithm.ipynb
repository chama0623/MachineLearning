```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

# メトロポリス法

```python
# サンプリングしたい分布
def f_tilde(z):
    return 1/(1 + z**2)

# メトロポリス法
# z0は初期値
def metropolis(z0):
    z = z0
    
    while True:
        z_star = z+random.uniform(-0.5,0.5)
        # 提案分布
        A = min(1,f_tilde(z_star)/f_tilde(z))
        if random.uniform(0,1)<A:
            z=z_star
        
        yield z
```

```python
sample = []
generator = metropolis(-5)
for i in range(1000000):
    _sample = next(generator)
    if -5<_sample and _sample<5:
        if random.uniform(0,1.0) <0.1:
            sample.append(_sample)

# 理想的な分布
x = np.linspace(-5,5,1000)
y = f_tilde(x)

fig,ax1 =  plt.subplots(facecolor="white")
ax2 = ax1.twinx()
ax1.hist(sample,bins=100)
ax2.plot(x,y,color="orange")

ax1.set_zorder(1)
ax2.set_zorder(2)
plt.show()
```
