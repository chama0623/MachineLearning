## Students't Distribution(studentのt分布) : 連続

確率密度関数
自由度nとする.
$$
f(x)=\frac{\Gamma \left( \frac{n+1}{2}\right)}{\sqrt{n \pi}\Gamma \left( \frac{n}{2} \right)} \left( 1+ \frac{x^2}{n}\right)^{-\left( \frac{n+1}{2} \right)}
$$

モーメント母関数 : **存在しない**

期待値 : $(n>0)$ $$E[X] = 0 $$ 分散 : $(n>2)$$$V[X] = \frac{n}{n-2}$$

共役事前分布 : なし  
再生性 : なし
他の分布との関係 :
- 確率変数$Z,U$がそれぞれN(0,1),$\chi^2_n$に従うとき,$T=\frac{Z}{\sqrt{\frac{U}{n}}}$は自由度nのt分布に従う.
- n=1のとき,コーシー分布である.

## 確率分布関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n=3 # degree of freedom

xscale = np.linspace(-5,5,100)

prob = stats.t.pdf(xscale,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

```python
nlist=[1,2,5,10] # degree of freedom
width=[-5,5] # 表示幅調整

xscalelist = []
problist = []
vlist = []
skewlist = []
kurtosislist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(width[0],width[1],100))
    problist.append(stats.t.pdf(xscalelist[i],nlist[i]))
    if nlist[i]>2:
        vlist.append(nlist[i]/(nlist[i]-2))
    else :
        vlist.append('not define')
    skewlist.append(stats.skew(problist[i]))
    kurtosislist.append(stats.kurtosis(problist[i]))
    
plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="Student's t("+str(nlist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.ylabel("P(X)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(0))
print("Variance : "+str(vlist))
print("Skewness : "+str(skewlist)) #歪度
print("Kurtosislist : "+str(kurtosislist)) #尖度
```

## 累積分布関数のグラフ

```python
n=4
xscale = np.linspace(-2,2,100)

prob = stats.t.pdf(xscale,n)
cdf = stats.t.cdf(xscale,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```

## Students't DistributionとStandard Normal Distributionの関係

```python
n=3 # degree of freedom

xscale = np.linspace(-5,5,100)

probt = stats.t.pdf(xscale,n)
probn = stats.norm.pdf(xscale,0,1)

plt.figure(facecolor="white")
plt.plot(xscale,probt,label="Student's t("+str(n)+")")
plt.plot(xscale,probn,label="N(0,1)")
plt.legend()
plt.show()
```
