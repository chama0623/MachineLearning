## Gamma distribution(ガンマ分布) : 連続

確率密度関数
$$
f(x)=\frac{1}{\Gamma(\alpha)}\frac{1}{\beta}\left( \frac{x}{\beta} \right)^{\alpha-1}e^{-\frac{x}{\beta}}
$$
ただし$\alpha>0,\beta>0,x>0$

モーメント母関数
$$
M_X(t) = \frac{1}{(1 - \beta t)^{\alpha}}
$$

期待値$$E[X] = \alpha \beta$$ 分散$$V[X] = \alpha \beta^2$$
他の分布との関係 :
- Ga(1,$\frac{1}{\lambda}$)はEx($\lambda$)である.
- Ga($\frac{n}{2}$,2)は$\chi^2_n$である.
- ポアソン分布の共役事前分布である.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
alpha = 2
beta = 3
xscale = np.linspace(0,15,100)

prob = stats.gamma.pdf(xscale,a=alpha,scale=beta)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

```python
alphalist=[0.5,2,4]
betalist =[3,6,6]

xscalelist = []
problist = []
mulist = []
vlist = []
for i in range(len(alphalist)):
    xscalelist.append(np.linspace(0,15,100))
    problist.append(stats.gamma.pdf(xscalelist[i],a=alphalist[i],scale=betalist[i]))
    mulist.append(alphalist[i]*betalist[i])
    vlist.append(alphalist[i]*betalist[i]*betalist[i])
    
plt.figure(facecolor="white")
for i in range(len(alphalist)):
    plt.plot(xscalelist[i],problist[i],label="Ga("+str(alphalist[i])+","+str(betalist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.ylabel("P(X)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

## 累積分布関数のグラフ

```python
alpha = 2
beta = 3
xscale = np.linspace(0,15,120)

prob = stats.gamma.pdf(xscale,a=alpha,scale=beta)
cdf = stats.gamma.cdf(xscale,a=alpha,scale=beta)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```

## Gamma関数と$\chi^2$分布, Exponential分布の対応

### Gamma関数と$\chi^2$分布

$\chi^2_n$は$Ga(\frac{n}{2},2)$に対応する.

```python
n = 3
xscale = np.linspace(0,15,100)

prob1 = stats.gamma.pdf(xscale,a=n/2,scale=2)
prob2 = stats.chi2.pdf(xscale,n)

plt.figure(facecolor="white")
plt.plot(xscale,prob1,label="Ga("+str(n)+"/2,2)")
plt.plot(xscale,prob2,linestyle="dashed",label="$\chi^2$"+str(n))
plt.legend()
plt.title("確率密度関数")
plt.show()
```

### Gamma分布とExponential分布
$Ex(\lambda)$は$Ga(1,\frac{1}{\lambda})$に対応する.

```python
la = 3
xscale = np.linspace(0,2,100)

prob1 = stats.gamma.pdf(xscale,a=1,scale=1/la)
prob2 = stats.expon.pdf(xscale,scale=1/la)

plt.figure(facecolor="white")
plt.plot(xscale,prob1,label="Ga(1,1/"+str(la)+")")
plt.plot(xscale,prob2,linestyle="dashed",label="Ex("+str(la)+")")
plt.legend()
plt.title("確率密度関数")
plt.show()
```
