## Discrete Uniform Distribution(離散一様分布) : 離散

確率密度関数
$x = 1,2, ... ,N$とする.
$$
f(x)=\frac{1}{N}
$$

期待値$$E[X] = \frac{N+1}{2}$$  分散$$V[X] = \frac{(N+1)(N-1)}{12}$$

共役事前分布 : なし  
使用例 : サイコロを投げたとき,各目がでる確率Xは離散一様分布(6)に従う.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
N=6

xscale = range(0,N)
prob = [1/N]*N

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("離散一様分布(N="+str(N)+")",fontsize=15)
plt.show()
```

## 累積分布関数のグラフ

```python
N=6

xscale = range(0,N)
prob = [1/N]*N
cdf = []
for i in range(N):
    cdf.append(sum(prob[:i+1]))
    
plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.plot(xscale,cdf)
plt.scatter(xscale,prob)
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("離散一様分布(N="+str(N)+")",fontsize=15)
plt.show()
```
