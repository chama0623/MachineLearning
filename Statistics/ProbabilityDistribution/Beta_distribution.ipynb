## Beta Distribution(ベータ分布) : 連続

確率密度関数
$$
f(x)= \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}
$$

期待値$$E[X] = \frac{\alpha}{\alpha+\beta}$$ 分散$$V[X] = \frac{\alpha \beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$$

他の分布との関係 : 
二項分布の共役事前分布である.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
alpha,beta = 54,50
xscale = np.linspace(0,1,100)

prob = stats.beta.pdf(xscale,alpha,beta)
plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()

# 解釈 : 表率xが不明であるコインを何回か投げたとき表がm,裏がn回出る予測値はBe(m+1,n+1)に従う
```

```python
alphalist=[2,2,2]
betalist=[2,4,6]
xscalelist = []
mulist = []
vlist = []
problist = []
for i in range(len(alphalist)):
    xscalelist.append(np.linspace(0,1,100))
    problist.append(stats.beta.pdf(xscalelist[i],alphalist[i],betalist[i]))
    mulist.append(alphalist[i]/(alphalist[i]+betalist[i]))
    vlist.append((alphalist[i]*betalist[i])/((alphalist[i]+betalist[i])**2*(alphalist[i]+betalist[i]+1)))
    
plt.figure(facecolor="white")
for i in range(len(alphalist)):
    plt.plot(xscalelist[i],problist[i],label="Be("+str(alphalist[i])+","+str(betalist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
```

```python
alpha,beta = 54,50
xscale = np.linspace(0,1,100)

prob = stats.beta.pdf(xscale,alpha,beta)
cdf = stats.beta.cdf(xscale,alpha,beta)
plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.grid()
plt.show()
```
