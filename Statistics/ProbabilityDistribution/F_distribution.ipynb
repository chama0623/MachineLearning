## F Distribution(F分布) : 連続

確率密度関数
2つの確率変数S,Tについて,それぞれ自由度m,mの$\chi^2$分布に従うとする．このとき次に示すYはF分布$F_{m,n}$に従う.
$$
F = \frac{\frac{S}{m}}{\frac{T}{n}}
$$
$$
F(y) = \frac{\left( \frac{m}{n} \right)^{\frac{m}{2}}}{B \left( \frac{m}{2},\frac{n}{2} \right)}\frac{y^{\frac{m}{2}-1}}{\left( 1+\frac{m}{n}y \right)^{\frac{m+n}{2}}}
$$

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n,m = 8,6
xscale = np.linspace(0,10,100)

prob = stats.f.pdf(xscale,n,m)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.show()
```

```python
nlist = [8,15,3]
mlist = [6,3,7]
width=[0,10] # 表示幅調整

xscalelist = []
problist = []
mulist = []
vlist = []
skewlist = []
kurtosislist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(width[0],width[1],200))
    problist.append(stats.f.pdf(xscalelist[i],nlist[i],mlist[i]))
    mulist.append(mlist[i]/(mlist[i]-2))
    vlist.append((2*(mlist[i]**2)*(nlist[i]+mlist[i]-2))/(nlist[i]*((mlist[i]-2)**2)*(mlist[i]-4)))
    skewlist.append(stats.skew(problist[i]))
    kurtosislist.append(stats.kurtosis(problist[i]))
    
plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="F("+str(nlist[i])+","+str(mlist[i])+")")
    #plt.plot([mulist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("x",fontsize=13)
plt.ylabel("P(X)",fontsize=13)
plt.legend()
plt.show()
print("Expected value : "+str(mulist))
print("Variance : "+str(vlist))
print("Skewness : "+str(skewlist)) #歪度
print("Kurtosislist : "+str(kurtosislist)) #尖度
```

## 累積分布関数のグラフ

```python
n,m = 6,8
xscale = np.linspace(0,10,100)

prob = stats.f.pdf(xscale,n,m)
cdf = stats.f.cdf(xscale,n,m)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.plot(xscale,cdf,label="累積分布関数")
plt.legend()
plt.grid()
plt.show()
```
