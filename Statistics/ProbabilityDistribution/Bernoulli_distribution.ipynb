## Bernoulli Distribution(ベルヌーイ分布) : 離散

確率pで成功する事象を1回試行したときの成功回数Xが従う分布

確率密度関数
$$
f(x)=p^k(1-x)^{1-x}
$$

積率母関数
$$
G_X(s) = sp+1-p
$$

期待値$$E[X] = p$$ 分散$$V[X] = p(1-p)$$

共役事前分布 : Beta分布  
再生性 : あり  
他の分布との関係 : Binomial distributionのn=1のときの分布である.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
p = 0.8
xscale = np.linspace(0,1,2)

prob = stats.bernoulli.pmf(xscale,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Bernoulli("+str(p)+")",fontsize=15)
plt.show()
```

## 累積分布関数のグラフ

```python
p = 0.2
xscale = np.linspace(0,1,2)

prob = stats.bernoulli.pmf(xscale,p)
prob_cd = stats.bernoulli.cdf(xscale,p)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Bernoulli("+str(p)+")",fontsize=15)
plt.legend()
plt.show()
```
