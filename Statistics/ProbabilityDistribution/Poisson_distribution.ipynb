## Poisson Distribution(ポアソン分布) : 離散

ある時間に平均$\lambda$回起きる事象が,ある時間に起きる回数Xが従う分布

確率密度関数
$$
f(x) = \frac{e^{-\lambda}\lambda^x}{x!}
$$

積率母関数
$$
G_X(s) = e^{\lambda(s-1)}
$$

期待値$$E[X] = \lambda$$ 分散$$V[X] = \lambda$$

共役事前分布 : Gamma分布  
再生性 : あり  
他の分布の関係 : 
- $\lambda$が十分大きいとき,正規分布N($\lambda$,$\lambda$)で近似できる.
- ある事象が起きてから次に起きるまでの時間は指数分布Ex($\lambda$)に従う.

## 確率密度関数のグラフ

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
la=4 # lambda 
n=15 # 確率分布のパラメータではなく表示する試行回数

xscale = np.linspace(0,n,n+1)
prob = stats.poisson.pmf(xscale,la)

plt.figure(facecolor="white")
plt.plot(xscale,prob)
plt.scatter(xscale,prob)
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Po("+str(la)+")",fontsize=15)
plt.show()
```

```python
lalist=[4,6,8]
nlist=[10,15,20]

xscalelist = []
problist = []
skewlist = []
kurtosislist = []
for i in range(len(nlist)):
    xscalelist.append(np.linspace(0,nlist[i],nlist[i]+1))
    problist.append(stats.poisson.pmf(xscalelist[i],lalist[i]))
    skewlist.append(stats.skew(problist[i]))
    kurtosislist.append(stats.kurtosis(problist[i]))

plt.figure(facecolor="white")
for i in range(len(nlist)):
    plt.plot(xscalelist[i],problist[i],label="Po("+str(lalist[i])+")")
    plt.scatter(xscalelist[i],problist[i])
    #plt.plot([lalist[i]]*2,[0,1],color="red") # if plot expected value
    
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
print("Lambda : "+str(lalist))
print("Skewness : "+str(skewlist)) #歪度
print("Kurtosislist : "+str(kurtosislist)) #尖度
```

## 累積分布関数のグラフ

```python
la=2
n=6
xscale = np.linspace(0,n,n+1)

prob = stats.poisson.pmf(xscale,la)
prob_cd = stats.poisson.cdf(xscale,la)

plt.figure(facecolor="white")
plt.plot(xscale,prob,label="確率密度関数")
plt.scatter(xscale,prob)
plt.plot(xscale,prob_cd,label="累積分布関数")
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.title("Po("+str(la)+")",fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

## Binomial distributionをPoisson distributionで近似する

```python
n,p = 20,0.1 # parameter of Binomial distribution
la=2 # parameter of Poisson distribution
nPo=6 # 確率分布のパラメータではなく表示する試行回数

xscaleBi = np.linspace(0,n,n+1)
xscalePo = np.linspace(0,nPo,nPo+1)
probBi = stats.binom.pmf(xscaleBi,n,p)
probPo = stats.poisson.pmf(xscalePo,la)

plt.figure(facecolor="white")
plt.plot(xscaleBi,probBi,label="Bi("+str(n)+","+str(p)+")")
plt.scatter(xscaleBi,probBi)
plt.plot(xscalePo,probPo,label="Po("+str(la)+")")
plt.scatter(xscalePo,probPo)
plt.xlim(0,nPo)
plt.xlabel("回数k",fontsize=13)
plt.ylabel("P(X=k)",fontsize=13)
plt.legend()
plt.show()
```
