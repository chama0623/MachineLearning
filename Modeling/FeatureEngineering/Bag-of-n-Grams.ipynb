```python
import pandas as pd
import numpy as np
import json
```

## Bag-of-n-Grams

### load data

```python
with open("Data/yelp_academic_dataset_review.json",encoding="utf-8") as f:
    js = []
    for i in range(10000): # 最初の10000件を取得
        js.append(json.loads(f.readline()))
review_df = pd.DataFrame(js)
review_df.head()
```

### Bag-of-n-Gramsの計算

```python
from sklearn.feature_extraction.text import CountVectorizer

bow_converter = CountVectorizer(token_pattern=r'\b\w+\b') # Bow
bigram_converter = CountVectorizer(ngram_range=(2,2),token_pattern=r'\b\w+\b') # Bigram
trigram_converter = CountVectorizer(ngram_range=(3,3),token_pattern=r'\b\w+\b') # Trigram
```

```python
# BoW
bow_converter.fit(review_df["text"])
words = bow_converter.get_feature_names()
print(words[:1000])
```

```python
bow_converter = CountVectorizer(token_pattern=r'\b\w+\b') # Bow
bow_converter.fit(review_df["text"])
words = bow_converter.transform(review_df["text"])
counts = np.array(words.sum(axis=0))[0]

for word,count in zip(bow_converter.get_feature_names(),counts):
    print(word,count)
```

```python
# Bigram
bigram_converter.fit(review_df["text"])
bigrams = bigram_converter.get_feature_names()
print(bigrams[:1000])
```

```python
# Trigram
trigram_converter.fit(review_df["text"])
trigrams = trigram_converter.get_feature_names()
print(trigrams[:1000])
```

## Stopwordsによる単語除去

### CountVectorizerでstopwordsの指定

```python
# 例として「able」をstopwordに指定する
bow_converter = CountVectorizer(token_pattern=r'\b\w+\b',stop_words=["able"])
bow_converter.fit(review_df["text"])
words = bow_converter.get_feature_names()

# 「able」がwordsに含まれるか
print("able" in words)
```

### nltkのstopwordsを用いた単語除去

```python
import nltk
from nltk.corpus import stopwords
```

```python
# nltkのstopwords

nltk.download('stopwords')
stop_words = stopwords.words('english')
print(stop_words)
```

```python
# 先に計算したBowからstopwwordsを除去してみる
print("len of words",len(words))
words_without_stopwords = [w for w in words if w not in stop_words] # stopwordsを除去した単語のリスト
deleted_pwords = [w for w in words if w in stop_words] # 除去された単語のリスト
print("len of words",len(words_without_stopwords))

words_without_stopwords[:1000]
```

## 頻度に基づく単語除去
出現頻度の高い単語, レアな単語を除去する. min_df,max_dfは小数(0～1)で指定するとdf値で除去する単語を決定する. 整数で与えた場合は出現頻度に基づいて単語を除去する.

```python
bow_converter = CountVectorizer(token_pattern=r'\b\w+\b',min_df=0.2,max_df=0.8)
bow_converter.fit(review_df["text"])
words = bow_converter.get_feature_names()
```
